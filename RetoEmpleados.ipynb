Ahora debes hacer la evaluación de las variables para quedarte con las mejores. Calcula la correlación lineal de cada una de las variables con respecto al Attrition.

Selecciona solo aquellas variables que tengan una correlación mayor o igual a 0.1, dejándolas en otro frame llamado EmpleadosAttritionFinal. No olvides mantener la variable de salidaAttrition; esto es equivalente a borrar las que no cumplen con el límite.

```python
# Calcular la matriz de correlación
correlation_matrix = empleadosattrition.corr()

# Filtrar las variables cuya correlación con Attrition sea mayor o igual a 0.1
attrition_correlation = correlation_matrix['Attrition_Yes'].abs()  # Usamos el valor absoluto para evitar signos negativos
selected_columns = attrition_correlation[attrition_correlation >= 0.1].index


# De empleadosattrition me vas a sacar solo las selected_columns
empleadosattrition_final = empleadosattrition[selected_columns]
```

Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal. Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucción EmpleadosAttritionPCA[:,0]).

Agrega el mínimo número de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal. Puedes usar la instrucción assign, columna por columna, llamando a cada unaC0, C1, etc., hasta las que vayas a agregar.

```python
from sklearn.decomposition import PCA

# Normalizar los datos antes de aplicar PCA (es recomendable para PCA)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(empleadosattrition_final)

# Aplicar PCA para explicar el 80% de la varianza
pca = PCA(n_components=0.80)  # Esto selecciona automáticamente el número de componentes para explicar el 80% de la varianza
pca_components = pca.fit_transform(scaled_data)


# Crear un DataFrame con los componentes principales
pca_df = pd.DataFrame(pca_components, columns=[f'C{i}' for i in range(pca_components.shape[1])])
pca_df.head(10)



```

```python
# Agregar los componentes principales al DataFrame original de empleadosattrition_final
empleadosattrition_final = empleadosattrition_final.reset_index(drop=True)  # Asegurarse de que el índice esté alineado
empleadosattrition_final = pd.concat([empleadosattrition_final, pca_df], axis=1)
empleadosattrition_final.head(10)
```
